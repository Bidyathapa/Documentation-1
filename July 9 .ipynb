# Equation of Progressive Wave 

##### The equation of progressive wave is given by:
$ y  =  a  sin  (  \omega t - kx )    $

where, $a$ is the amplitude, $\omega$ is the angular frequency, k is the the constant, x and t are variables.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#y=asin(t)
t=np.arange(1,10,0.01)

a=10 #in mm
y=a*np.sin(t)
z=a*np.cos(t)
plt.figure(figsize=(10,5))
plt.plot(t,y,'-.',color='r',lw=3)
plt.plot(t,z,'-.',color='g',lw=3)
plt.xlabel('time(s)',fontsize=14)
plt.ylabel('displacement(mm)',fontsize=14)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0)
plt.savefig('shm.jpg')
```
