```python
import numpy as np
import matplotlib.pyplot as plt
#y=asin(t)
t=np.arange(1,10,0.01)

a=10 #in mm
y=a*np.sin(t)
plt.plot(t,y,'-.',color='r',lw=3)
plt.xlabel('time(s)',fontsize=14)
plt.ylabel('displacement(mm)',fontsize=14)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0)
```
