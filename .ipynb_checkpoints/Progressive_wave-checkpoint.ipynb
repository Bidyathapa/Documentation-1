# Stationary Wave
Stationary wave is formed when two progressive travelling from opposite direction with equal velocity are met at a place. 
### Stationary Water Wave
![img](https://dr282zn36sxxg.cloudfront.net/datastreams/f-d%3Ae634bea6e8d1730e176db9ea0177206a7b46e97c6ba8286418880fba%2BIMAGE_TINY%2BIMAGE_TINY.1)


#### A simple progressive wave can be plotted using jupyter notebook as shown below:

```python
import numpy as np
import matplotlib.pyplot as plt

#y=asin(t)
t=np.arange(1,100,0.01)

a=10 #in mm
y=a*np.sin(t)
plt.plot(t,y,'-',color='r',lw=3)
plt.xlabel('time(s)',fontsize=14)
plt.ylabel('displacement(mm)',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0,lw=3,color='blue')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#y=asin(t)
t=np.arange(1,10,0.001)

a=10 #in mm
y=a*np.sin(t)
plt.plot(t,y,'.',color='r',lw=3)
plt.xlabel('time(s)',fontsize=20,color='blue')
plt.ylabel('displacement(mm)',fontsize=20, color='blue')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0,lw=3,color='g')
```

```python
#y=asin(wt)
t=np.arange(1,10,0.001)
w=2*np.pi*1000 #in rad/s
a=10 #in mm
y=a*np.sin(t)
z=a*np.cos(t)
plt.plot(t,y,'-.',color='r',lw=3,label='Sine Function' )
plt.plot(t,z,'-',color='green',lw=3,label='Sine Function')
plt.xlabel('time(s)',fontsize=20,color='blue')
plt.ylabel('displacement(mm)',fontsize=20, color='blue')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0,lw=3,color='g')
plt.legend(fontsize=16,loc='best')
```

```python
#y1=a1sin(wt-kx), y2=a2sin(wt+kx)
t=np.arange(1,5,0.001)
x=np.arange(1,5,0.001)
w=10 #in rad/s
a1=10 #in mm
a2=10
k=2 
y1=a1*np.sin(w*t-k*x)
y2=a2*np.sin(w*t+k*x)
plt.plot(t,y1,'-.',color='r',lw=3,label='Forward Wave')
plt.plot(t,y2,'-',color='green',lw=3,label='Reverse Wave')
plt.xlabel('time(s)',fontsize=20,color='blue')
plt.ylabel('displacement(mm)',fontsize=20, color='blue')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0,lw=3,color='blue')
plt.legend(fontsize=16,loc='best')
```
